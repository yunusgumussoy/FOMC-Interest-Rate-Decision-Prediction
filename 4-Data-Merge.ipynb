```python
import pandas as pd

# FOMC-date-decision-data
# Read the Excel file into a DataFrame
fomc_data = pd.read_csv('FOMC-date-decision-data.csv')

# Display the first few rows
fomc_data.head()
```

```python
# Economic Index and Indicators Data 
# Read the Excel file into a DataFrame
data = pd.read_csv('data.csv')

# Display the first few rows
data.head()
```

```python
# Rename column
data.rename(columns={'Unnamed: 0': 'Release Date'}, inplace=True)
```

```python
data.info()
```

```python
data.tail()
```

```python
# FOMC Statements and Minutes Data
# Read the Excel file into a DataFrame
statement_data = pd.read_csv('FOMC-Statement-NLP.csv')

# Display the first few rows
statement_data.head()
```

```python
# Drop redundant columns
statement_data.drop(columns=["Unnamed: 0", "Announcement Date"], inplace=True)
```

```python
statement_data.info()
```

```python
statement_data.tail(9)
```

```python
# Merge

# Convert date columns to datetime format
statement_data["Release Date"] = pd.to_datetime(statement_data["Release Date"])
fomc_data["Release Date"] = pd.to_datetime(fomc_data["Release Date"])
data["Release Date"] = pd.to_datetime(data["Release Date"])
```

```python
# Merge the two datasets based on dates
first_merged = pd.merge(
    statement_data,
    fomc_data,
    left_on="Release Date",
    right_on="Release Date",
    how="left"
)
```

```python
first_merged.head()
```

```python
first_merged.info()
```

```python
first_merged.tail()
```

```python
# Merge the two datasets based on dates
final_merged = pd.merge(
    first_merged,
    data,
    left_on="Release Date",
    right_on="Release Date",
    how="left"
)
```

```python
final_merged.info()
```

```python
final_merged.head()
```

```python
# Sort the final dataset by Release Date for clarity
final_merged = final_merged.sort_values(by='Release Date')
```

```python
# Drop redundant date column from the second dataset after the merge
final_merged.drop(columns=["Time","Forecast"], inplace=True)

# Display the merged DataFrame
print(final_merged)
```

```python
# Save to CSV
final_merged.to_csv('final-data.csv')
print("Data saved to final-data.csv")
```
